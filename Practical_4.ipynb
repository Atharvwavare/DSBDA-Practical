```python
#a. Creating a subset by selecting specific columns by their name
subset1 = df[['Type','Category','Post Weekday','Paid','Total Interactions']]
subset1
```

```python
subset1.shape
```

```python
# Creating a subset by filtering a value of row
subset2 = df[df['Post Hour']>4]
subset2
```

```python
subset2.shape
```

```python
# Creating a subset by selecting specific rows and columns
subset3 = df.loc[df['Lifetime Engaged Users']>600,['Post Month','Post Weekday']]
subset3
```

```python
subset3.shape
```

```python
# Creating a subset by index range
subset4 = df.iloc[5:10]
subset4
```

```python
subset4.shape
```

```python
# Creating a subset based on multiple conditions
subset5 = df[(df['Lifetime Post Total Reach']>10000)&(df['Post Hour']>5)]
subset5
```

```python
subset5.shape
```

```python
# Creating a subset by selecting rows
submerge1=df[['share','Paid']].loc[0:5]
submerge1
```

```python
submerge1.shape
```

```python
submerge2 = df[['Paid','Lifetime Post Consumers']].loc[4:8]
submerge2
```

```python
submerge2.shape
```

```python
# Merging data using inner join
merge_inner=pd.merge(submerge1,submerge2,on='Paid',how='inner')
merge_inner
```

```python
# Merging data using inner join
merge_outer=pd.merge(submerge1,submerge2,on='Paid',how='outer')
merge_outer
```

```python
# Sorting by one column in ascending order
sort_single_asc=df.sort_values(by='Total Interactions',ascending=True)
sort_single_asc
```

```python
# Sorting by one column in descending order
sort_single_dsc=df.sort_values(by='Total Interactions',ascending=False)
sort_single_dsc
```

```python
# Sorting by multiple columns, specifying ascending or descending for each
sort_multiple=df.sort_values(by=['Paid','Total Interactions'],ascending=[True,False])
sort_multiple
```

```python
# Sorting data by row index
sort_index=df.sort_index(axis=0,ascending=False)
sort_index
```
