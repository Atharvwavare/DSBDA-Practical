```python
# a data cleaning

# Check for null values
null_values = df.isnull().sum()
null_values
```

```python
# Check for data types
data_types = df.dtypes
data_types
```

```python
# b. data integration

df1 = df[['age','sex','cp','trtbps']]
df2 = df[['oldpeak','slp','caa']]
df_merge = pd.concat([df1,df2],axis=1).drop_duplicates().reset_index(drop=True)
df_merge
```

```python
 # c. Data transformation
from sklearn.preprocessing import MinMaxScaler
# Normalizing selected numerical features
scaler = MinMaxScaler()
df[['age', 'trtbps', 'chol', 'thalachh', 'oldpeak']] = scaler.fit_transform(df[['age', 'trtbps', 'chol', 'thalachh', 'oldpeak']])
df
```

```python
# d) Error Correcting
df.columns
```

```python

# Fixing any values outside expected ranges
df_sex =  df['sex'].apply(lambda x: 1 if x == 1 else 0)  # Ensuring binary
df_chol = df[df['chol'] > 0.28]  # Removing invalid cholesterol values
df_sex
df_chol

```

```python
# e. Data model building
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Splitting dataset
X = df.drop("output", axis=1)
y = df["output"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Building model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Prediction and accuracy
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
